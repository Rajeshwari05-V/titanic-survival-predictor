```python
# Day 2: Pandas Basics Practice (M3)
# Rajeshwari's 90-Day AI/ML Journey ðŸš€
import pandas as pd
# --- Step 1: Create a Series ---
numbers=pd.Series([10,20,30,40,50])
print("Series:")
print(numbers)
print("/nAcessing Elements:")
print("First element:",numbers[0])
print("Last element:",numbers.iloc[-1])
# --- Step 2: Create a DataFrame ---
data={
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["Bangalore", "Mumbai", "Delhi"]
}
df=pd.DataFrame(data)
print("\nDataFrame:")
print(df)
# --- Step 3: Explore the DataFrame ---
print("\nHead:")
print(df.head())

print("\nInfo:")
print(df.info())

print("\nDescribe:")
print(df.describe())
# --- Step 4: Indexing & Selection ---
print("\nSelect single column:")
print(df["Name"])

print("\nSelect multiple columns:")
print(df[["Name", "City"]])

print("\nSelect row using loc (label):")
print(df.loc[0])

print("\nSelect row using iloc (index):")
print(df.iloc[1])
# --- Step 5: Filtering ---
print("\nFilter Age > 28:")
print(df[df["Age"]>28])
# --- Step 6: Save to CSV ---
df.to_csv("sample_people.csv", index=False)
print("\nData saved to sample_people.csv âœ…")
```
